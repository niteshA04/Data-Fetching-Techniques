# Importing Libraries
- Pandas will be used for handeling data.
- Requests will be used for fetcing data from api.

```python
import pandas as pd
import requests
```

# Storing the data
- In requests library there is a function called .get() which helps in gathering of data from API.
- I'm storing the data in a varible called req.

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1')
```

# Converting the web data into JSON

```python
response.json()
```

# As we only need results part 

```python
response.json()['results']
```

# Creating empty dataframe
- So that we can store the loop data into it

```python
df = pd.DataFrame()
```

```python
df
```

# Appling the loop for fetching the data for all pages
- As there are 428 pages of data on that website I'm witing 429
- I'm using range function to achieve this

```python
for i in range(1,429):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = df.append(temp_df,ignore_index=True)
```

```python
df.head(5)
```

# Exporting it in CSV format

```python
df.to_csv('top_rated_movies.csv')
```
